```python
# Function to scrape articles from The Daily Star
def scrape_the_daily_star(url):
    try:
        response = requests.get(url)
        response.raise_for_status()
    except requests.exceptions.RequestException as e:
        print(f"Error fetching the URL: {e}")
        return []

    soup = BeautifulSoup(response.text, 'html.parser')
    # Removed extra indentation here
    articles = soup.find_all('div', class_='content-details')
    article_texts = []
    for article in articles:
        title = article.find('h2').get_text(strip=True) if article.find('h2') else "No title"
        content = article.find('p').get_text(strip=True) if article.find('p') else "No content"
        article_texts.append(f"{title}\n{content}")

    return article_texts
```

```python
# Function to summarize articles using GPT-4
def summarize_text(text):
    try:
        response = openai.Completion.create(
            engine="text-davinci-003",
            prompt=f"Summarize the following text:\n\n{text}",
            max_tokens=150,
            temperature=0.5,
        )
        return response.choices[0].text.strip()
    except Exception as e:
        print(f"Error with OpenAI API: {e}")
        return "Summary could not be generated."
```

Web Scrapping and summerize multiple pages

```python
# Function to determine the total number of pages dynamically
def get_total_pages(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')

    # Example: find the total number of pages from pagination links
    pagination = soup.find('ul', class_='pagination')
    if pagination:
        pages = pagination.find_all('li')
        return int(pages[-2].get_text())  # Usually, the last page number is second to last in pagination
    return 1
```

```python
# Function to scrape and summarize multiple pages
def scrape_and_summarize(url, num_pages=1):
    all_summaries = []

    for page in range(1, num_pages + 1):
        page_url = f"{url}?page={page}"
        print(f"Scraping page {page}: {page_url}")
        articles = scrape_the_daily_star(page_url)

        for article in articles:
            summary = summarize_text(article)
            all_summaries.append(summary)
            print(f"Summary:\n{summary}\n")

        # Be respectful of the website's servers; add a short delay
        time.sleep(2)

    return all_summaries
```

```python
# Function to scrape and summarize multiple pages
def scrape_and_summarize(url, num_pages=1):
    all_summaries = []

    for page in range(1, num_pages + 1):
        page_url = f"{url}?page={page}"
        print(f"Scraping page {page}: {page_url}")
        articles = scrape_the_daily_star(page_url)

        # Check if articles is None before iterating
        if articles is not None:
            for article in articles:
                summary = summarize_text(article)
                all_summaries.append(summary)
                print(f"Summary:\n{summary}\n")
        else:
            print(f"No articles found on page {page}.")

        # Be respectful of the website's servers; add a short delay
        time.sleep(2)

    return all_summaries
```

```python
 # Example usage with dynamic page count

base_url = "https://www.thedailystar.net/politics"
total_pages = get_total_pages(base_url)
print(f"Total pages found: {total_pages}")

all_summaries = scrape_and_summarize(base_url, 5)
```
